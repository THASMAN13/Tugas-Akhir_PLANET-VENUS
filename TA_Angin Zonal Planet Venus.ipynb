```python
import numpy as np
import pandas as pd
import numpy as np

# Konstanta dan variabel
R = 6051.8  # Jari-jari Venus dalam km
h = 46      # Ketinggian rata-rata dalam km
sl = 360    # Satuan putaran lingkaran dalam derajat
Pr = 243.0225   # Periode Rotasi planet Venus dalam hari

# Membaca data dari file Excel
data =  pd.read_excel(r'C:/Users/hi/Downloads/WindZonalVenus2022/d1.xlsx')
data

from datetime import datetime

#T21
t21 = []

for i,j,k,l,m,n in zip(data['Lintang 1'], data['Bujur 1'], data['Lintang 2'], data['Bujur 2'], data['Jam ke 1'], data['Jam ke 2']):
  jam1 = datetime.strptime(m, "%H:%M:%S.%f")
  jam2 = datetime.strptime(n, "%H:%M:%S.%f")
  deltajam = jam2-jam1
  deltajam = deltajam.total_seconds() / 3600
  # Menghitung kecepatan angin zonal
  u = ((l - j) * (R + h) *np.cos(i/(180/np.pi)) / deltajam) * (2 * np.pi) / 360

  rel_vel_ven_l40 = (2 * (np.pi) * R / 360) * (360 / (Pr * 24)) * np.cos(i/(180/np.pi))
  rel_vel_ven = np.abs(u) - rel_vel_ven_l40
  eastwest_vel_ven = rel_vel_ven

  t21.append(eastwest_vel_ven*-1*(10/36))

  #print("Kecepatan rotasi Venus : ", ang_vel_ven_jam, "km/jam")
  print(" ")
  print("Kecepatan Angin Timur ke Barat Venus 1 Februari 2022 Saat T21 Lintang 40 : ", eastwest_vel_ven*-1 * 1000/3600, "m/s")
  
#T32
t32 = []

for i,j,k,l,m,n in zip(data['Lintang 2'], data['Bujur 2'], data['Lintang 3'], data['Bujur 3'], data['Jam ke 2'], data['Jam ke 3']):
  jam2 = datetime.strptime(m, "%H:%M:%S.%f")
  jam3 = datetime.strptime(n, "%H:%M:%S.%f")
  deltajam = jam3-jam2
  deltajam = deltajam.total_seconds() / 3600
  # Menghitung kecepatan angin zonal
  u = ((l - j) * (R + h) *np.cos(i/(180/np.pi)) / deltajam) * (2 * np.pi) / 360

  rel_vel_ven_l40 = (2 * (np.pi) * R / 360) * (360 / (Pr * 24)) * np.cos(i/(180/np.pi))
  rel_vel_ven = np.abs(u) - rel_vel_ven_l40
  eastwest_vel_ven = rel_vel_ven

  t32.append(eastwest_vel_ven*-1*(10/36))

  #print("Kecepatan rotasi Venus : ", ang_vel_ven_jam, "km/jam")
  print(" ")
  print("Kecepatan Angin Timur ke Barat Venus 1 Februari 2022 Saat T32 Lintang 40 : ", eastwest_vel_ven*-1 * 1000/3600, "m/s")  
    
#T43
t43 = []

for i,j,k,l,m,n in zip(data['Lintang 3'], data['Bujur 3'], data['Lintang 4'], data['Bujur 4'], data['Jam ke 3'], data['Jam ke 4']):
  jam3 = datetime.strptime(m, "%H:%M:%S.%f")
  jam4 = datetime.strptime(n, "%H:%M:%S.%f")
  deltajam = jam4-jam3
  deltajam = deltajam.total_seconds() / 3600
  # Menghitung kecepatan angin zonal
  u = ((l - j) * (R + h) *np.cos(i/(180/np.pi)) / deltajam) * (2 * np.pi) / 360

  rel_vel_ven_l40 = (2 * (np.pi) * R / 360) * (360 / (Pr * 24)) * np.cos(i/(180/np.pi))
  rel_vel_ven = np.abs(u) - rel_vel_ven_l40
  eastwest_vel_ven = rel_vel_ven

  t43.append(eastwest_vel_ven*-1*(10/36))

  #print("Kecepatan rotasi Venus : ", ang_vel_ven_jam, "km/jam")
  print(" ")
  print("Kecepatan Angin Timur ke Barat Venus 1 Februari 2022 Saat T43 Lintang 40 : ", eastwest_vel_ven*-1 * 1000/3600, "m/s")
    
    
#T54
t54 = []

for i,j,k,l,m,n in zip(data['Lintang 4'], data['Bujur 4'], data['Lintang 5'], data['Bujur 5'], data['Jam ke 4'], data['Jam ke 5']):
  jam4 = datetime.strptime(m, "%H:%M:%S.%f")
  jam5 = datetime.strptime(n, "%H:%M:%S.%f")
  deltajam = jam5-jam4
  deltajam = deltajam.total_seconds() / 3600
  # Menghitung kecepatan angin zonal
  u = ((l - j) * (R + h) *np.cos(i/(180/np.pi)) /  deltajam) * (2 * np.pi) / 360

  rel_vel_ven_l40 = (2 * (np.pi) * R / (Pr * 24)) * np.cos(i/(180/np.pi))
  rel_vel_ven = np.abs(u) - rel_vel_ven_l40
  eastwest_vel_ven = rel_vel_ven

  t54.append(eastwest_vel_ven*-1*(10/36))

  #print("Kecepatan rotasi Venus : ", ang_vel_ven_jam, "km/jam")
  print(" ")
  print("Kecepatan Angin Timur ke Barat Venus 1 Februari 2022 Saat T54 Lintang 40 : ", eastwest_vel_ven*-1 * 1000/3600, "m/s")
  
# menghitung kecepatan rata-rata angin harian
t=t21+t32+t43+t54
print("Kecepatan Rata-Rata Angin Timur ke Barat (Zonal) Planet Venus Pada 1 Februari 2022 Lintang 40 : ", np.average(t), "m/s")
```
